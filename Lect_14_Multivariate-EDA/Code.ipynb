### Lecture 14: EDA using Multivariate Analysis

##### Importing Libraries

```python
import pandas as pd
import seaborn as sns
```

##### Importing Useful Datasets

```python
titanic=sns.load_dataset('titanic')
tips=sns.load_dataset('tips')
flights=sns.load_dataset('flights')
```

```python
tips.head()
```

##### 1. (Numerical/Numerical)

a. ScatterPlot

```python
sns.scatterplot(x=tips.total_bill,y=tips.tip,hue=tips.sex,style=tips.day,size=tips.time)
```

b. LinePlot

```python
flights.head()
```

```python
new=flights.groupby('year')['passengers'].sum().reset_index()
```

```python
new
```

```python
sns.lineplot(x=new.year,y=new.passengers)
```

##### 2. (Numerical/Categorical)

a. Barplot

```python
sns.barplot(x=tips['time'],y=tips['tip'],hue=tips.sex)
```

```python
tips.head()
```

b. Boxplot

```python
sns.boxplot(x=tips.time,y=tips.total_bill,hue=tips.smoker)
```

##### 3. (Categorical/Categorical)

a. HeatMap

```python
titanic.head()
```

```python
sns.heatmap(pd.crosstab(titanic.pclass,titanic.survived))
```

```python
(titanic.groupby('pclass')['survived'].mean()*100).plot(kind='bar')
```

```python
(titanic.groupby('pclass')['survived'].mean()*100).plot(kind='pie',autopct='%.2f')
```

##### 4. Pairplot (All)

```python
tips.head()
```

```python
sns.pairplot(tips,hue='smoker')
```
