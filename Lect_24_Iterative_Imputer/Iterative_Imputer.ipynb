### Lecture 24: Handing Missing Values Iterative Imputer

Steps


1. Checking columns with Nan Values (works best with MAR)
2. Splitting data into training & testing
3. Making object of Iterative Imputer class
4. Applying fit_transform on Training data & transform on Test data
   
    ###### As after fit_transform X_train & X_test will be numpy array so if you want you can change them back to data frame (your choice) 

```python

import numpy as np
import pandas as pd
from sklearn.experimental import enable_iterative_imputer # enables the imputer
from sklearn.impute import IterativeImputer # imputer itself

from sklearn.model_selection import train_test_split 
from sklearn.linear_model import LinearRegression
```

```python
df=pd.read_csv('framingham.csv')
df.head(5)
```

Train Test Split

```python
X=df.drop(columns='TenYearCHD',axis=1)
Y=df.TenYearCHD
```

```python
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,random_state=42,test_size=0.2)
```

```python
X_train.sample(10)
```

Applying Iterative imputer

```python
imputer = IterativeImputer(
    estimator=LinearRegression(),  
    max_iter=100,
    random_state=42,
    verbose=2
)
```

```python
X_train.isnull().sum()
```

Applying Fit Transform

```python
X_train['education']=imputer.fit_transform(X_train[['education']])
X_test['education']=imputer.transform(X_test[['education']])

X_train['glucose']=imputer.fit_transform(X_train[['glucose']])
X_test['glucose']=imputer.transform(X_test[['glucose']])

# X_train=imputer.fit_transform(X_train)
# X_test=imputer.fit_transform(X_test)
```
