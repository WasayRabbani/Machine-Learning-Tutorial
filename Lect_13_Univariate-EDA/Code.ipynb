### Lecture 13: EDA using Univariate Analysis

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv('Automobile.csv')
```

```python
df.head()
```

```python
df['origin'].nunique()
```

#### Numerical Data

1. Histogram

```python
plt.hist(df['horsepower'],bins=10)
```

2. Box Plot

```python
sns.boxplot(df['horsepower'],orient='h')
```

3. Displot

```python
sns.displot(df['horsepower'],kde=True,stat='density')
```

#### Categorical Data

1. Pie Chart

```python
df['origin'].value_counts().plot(kind='pie',autopct='%.2f')
```

2. Count Plot

```python
sns.countplot(df['origin'])
```
