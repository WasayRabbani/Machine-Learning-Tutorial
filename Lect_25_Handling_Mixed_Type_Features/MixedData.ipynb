### Lecture 25: Handing Mixed Type Features

Steps
1. See which column has mixed variables
2. Checking total no of categories in a column
3. Checking type of Mixed Data (Single,multi,space between them)
4. Extracting Numerical Part 
5. Extracting Categorical Part

Importing Libraries

```python
import pandas as pd
import numpy as np
```

Importing Data

```python
df=pd.read_csv('avocado_ripeness_dataset.csv')
df.head()
```

Step 1: Checking which columns have mixed data

###### Here Tag & Quality are columns having mixed data

Step 2: Checking total no of categories in a column

```python
df['Quality'].unique()
```

Step 3: Check type of mixed data (Single/Multi)

#### Single Type 

Step 4: Extracting Numerical Part

```python
df['Tag_numerical']=pd.to_numeric(df['Tag'],downcast='integer',errors='coerce')
```

Step 5: Extracting Categorical Part

```python
df['Tag_categorical']=np.where(df['Tag_numerical'].isnull(),df['Tag'],np.nan)
```

```python
df.sample(5)
```

#### Multi No Space

```python
df['Quality'].unique()
```

Step 4: Capture Numerical Part

```python
df['Quality_numerical'] = df['Quality'].str.extract(r'(\d+)') 
```

Step 5: Capturing Categorical Part

```python
df['Quality_categorical'] = df['Quality'].str[0] 
```

```python
df.head()
```

#### Multi with Space 

```python
df['Avocado Num'].unique()
```

Step 4: Capture Numerical Part

```python
# Getting numerical part as string data type
df['Avocado Num Numerical']=df['Avocado Num'].apply(lambda f:f.split()[-1])
```

```python
# Converting string type to numerical type
df['Avocado Num Numerical']=pd.to_numeric(df['Avocado Num Numerical'],errors='coerce',downcast='integer')
```

Step 5: Capture Categorical Part

```python
# Splitting num and cat part
df['Avocado Num Cat']=df['Avocado Num'].apply(lambda f:f.split()[0]) 

# Making cat & num columns
df['Avocado Num Cat']=np.where(df['Avocado Num Cat'].str.isdigit(),np.nan,df['Avocado Num Cat'])
```
