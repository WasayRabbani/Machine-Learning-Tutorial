Importing Libraries

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import OneHotEncoder
```

Importing Data

```python
df=pd.read_csv('car_price_prediction.csv')
df.head()
```

```python
df=pd.read_csv('car_price_prediction.csv',usecols=['Model','Gear box type','Manufacturer','Price'])
```

Setting Manufacturer Column & Setting Model Column

```python
count=df['Manufacturer'].value_counts()
rare_manufacturers=count[count<2000].index
df['Manufacturer']=df['Manufacturer'].replace(rare_manufacturers,'Other' )

count_model=df['Model'].value_counts()
rare_models=count_model[count_model<1000].index
df['Model']=df['Model'].replace(rare_models,'Others' )
```

Step 2: Splitting Data into Training & Testing Data

```python
X=df.drop(columns='Price',axis=1)
Y=df.Price
```

```python
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.2,random_state=42)
```

Step 3: Making object of OHE class


```python
OHE=OneHotEncoder(sparse_output=False,handle_unknown='ignore',dtype='int')
```

Step 4:Applying fit_transform and save to new variable

```python
X_train_new=OHE.fit_transform(X_train[['Manufacturer','Model','Gear box type']])
X_test_new=OHE.transform(X_test[['Manufacturer','Model','Gear box type']])
```

```python
col=OHE.get_feature_names_out(['Manufacturer','Model','Gear box type'])
```

```python
# Update X train & X test
X_train=pd.DataFrame(X_train_new,columns=col)
X_test=pd.DataFrame(X_test_new,columns=col)
```
